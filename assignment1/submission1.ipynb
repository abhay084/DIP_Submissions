```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('office.jpg')
```

```python
R,G,B = img[:,:,0],img[:,:,1],img[:,:,2]
```

```python
gray_data = (R+G+B)/3
```

```python
plt.imshow(gray_data,cmap='gray')
```

```python
binary_data = np.zeros(gray_data.shape)
```

```python
binary_data
```

```python
binary_data.shape
```

```python
for i in range(1080):
    for j in range(1920):
        if gray_data[i,j]>50:
            binary_data[i,j]=1
        else:
            binary_data[i,j]=0
```

```python
plt.imshow(binary_data,cmap='gray')
```

```python
binary_data
```

```python
img1 = gray_data + binary_data
```

```python
plt.imshow(img1,cmap='gray')
```

```python
img2 = img1 + 20
```

```python
plt.imshow(img2,cmap='gray')
```
