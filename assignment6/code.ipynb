```python
import matplotlib.pyplot as plt
import matplotlib.image as mping
```

```python
# read image 
img = mping.imread('img.jpg')
```

```python
# original image
plt.imshow(img)
plt.show()
```

```python
import numpy as np
import math
```

```python
np.shape(img)
```

```python
# generating increased cconstrast image
for i in range(675):
    for j in range(1200):
        for k in range(3):
            if(img[i][j][k]<128):
                img[i][j][k]=img[i][j][k]*2
            else:
                img[i][j][k]=math.floor(img[i][j][k]*1.5)
                
```

```python
# increased constrast image
plt.imshow(img)
plt.show()
```

```python
#  generating decreased constrast image
for i in range(675):
    for j in range(1200):
        for k in range(3):
            if(img[i][j][k]<128):
                img[i][j][k]=img[i][j][k]*1.5
            else:
                img[i][j][k]=math.floor(img[i][j][k]*0.9)
                
```

```python
plt.imshow(img)
plt.show()
```
