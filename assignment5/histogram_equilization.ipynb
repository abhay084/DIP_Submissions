```python
import matplotlib.pyplot as plt
```

```python
img = plt.imread('nature.jpg')
```

```python
img
```

```python
B,G,R = img[:,:,0],img[:,:,1],img[:,:,2]
```

```python
B
```

```python

gray_img = R*(0.2989) + G*(0.5870) + B*(0.1140)  
```

```python
gray_img
```

```python
gray_img.shape
```

```python
for i in range(225):
    for j in range(401):
        gray_img[i,j] = int(round(gray_img[i,j]))
```

```python
gray_img
```

```python

type(gray_img[i,j])
```

```python
gray_img = gray_img.astype(int)
```

```python
gray_img
```

```python
pixel_count = [0]*256
```

```python
gray_img.shape
```

```python
for i in range(225):
    for j in range(401):
        pixel_count[gray_img[i,j]]= pixel_count[gray_img[i,j]] + 1
```

```python
pixel_count
```

```python
prob_count = [0]*256
```

```python
n=0
for i in pixel_count:
    n=n+i
```

```python
for i in range(256):
    prob_count[i] = pixel_count[i]/n
```

```python
prob_count
```

```python
cdf = [0]*256
```

```python
cdf[0]=pixel_count[0]
for i in range(1,256):
    cdf[i]=pixel_count[i]+cdf[i-1]
```

```python
cdf
```

```python
histo_equi = [0]*256
```

```python
for i in range(256):
    histo_equi[i] = ((cdf[i]-cdf[0])/((225*401)-cdf[0]))*255 +1
```

```python
histo_equi
```

```python
plt.plot(histo_equi)
plt.show()
```

```python
new_img=gray_img
```

```python
for i in range(225):
    for j in range(401):
        for k in range(256):
            if gray_img[i,j]==k:
                new_img[i,j]=histo_equi[k]
                
                
```
